```python
import numpy as np
import pandas as pd
from statsmodels.stats.anova import anova_lm
```

```python
df = pd.read_excel('/content/MLR-House Prize (1).xlsx')
```

```python
df[:10]
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

#performing three-way ANOVA
model = ols("""Y ~ C(X1) + C(	X2) + C(X3) +
               C(X1):C(X2) + C(X1):C(X3) + C(X2):C(X3) +
               C(X1):C(X2):C(X3)""", data=df).fit()
```

```python
anova_table=anova_lm(model, typ=3)
anova_table
```
